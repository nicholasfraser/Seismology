```python
%matplotlib inline
!pip install obspy
import obspy
from obspy.clients.fdsn import Client
from obspy import UTCDateTime

#!pip install cartopy
#import cartopy
```

```python
client = Client("IRIS")

t = UTCDateTime("2016-09-09 00:30:00") # DPRKS
[ sbef, saft ] = [ 0, 240 ]

st = client.get_waveforms("IC", "MDJ", "00", "BH*", t-sbef, t+saft, attach_response = True)
st.remove_response(output='VEL')
st.plot(linewidth = 0.75, equal_scale = False)
```

```python
# something from the New Mexico sensor

t = UTCDateTime("2010-02-27T06:45:00.000")
st = client.get_waveforms("IU", "ANMO", "00", "LHZ", t, t + 60 * 60)
st.plot()  
```

```python
# this with minimum magnitude of 9 gets the Indian Ocean event in Dec 2004 and the Fukushima earthquake in Mar 2011

starttime = UTCDateTime("2002-01-01")
endtime = UTCDateTime("2022-09-09")
cat = client.get_events(starttime=starttime, endtime=endtime, minmagnitude=9, catalog="ISC")
print(cat.__str__(print_all=True))
```
